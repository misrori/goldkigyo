```python
import os
data_files = [x for x in os.listdir('/content/flights/csv_data') if x.endswith('BUD')]
data_files.sort()
data_files
```

```python
x = data_files[0]
# convert to date
```

```python
import pandas as pd
df_list = []
for x in data_files:
    tdf = pd.read_csv('/content/flights/csv_data/' + x)
    tdf['filename'] = x
    tdf['data_date'] = pd.to_datetime(x.replace('_BUD', ''), format='%Y_%m_%d')
    df_list.append(tdf)

df = pd.concat(df_list)
df.reset_index(drop=True, inplace=True)

```

```python
df.dtypes
```

```python
# id of flight
df['flight_id'] = df['to'] + ' - ' +df['dest_id'] +df['outgoing_start']+df['incomming_arrival']
df.head()
```

```python
# visualize the price change for one id
row_id = df['flight_id'][1570]
# filter df to one flight
one_row_df = df[df['flight_id']==row_id ]
```

```python
import plotly.express as px

row_id = df['flight_id'][45870]
# filter df to one flight
one_row_df = df[df['flight_id']==row_id ]


plot_title = one_row_df['to'].iloc[0] + ' ' + one_row_df['dest_id'].iloc[0] + ' ' + one_row_df['outgoing_start'].iloc[0] + ' - ' + one_row_df['incomming_arrival'].iloc[0]
p = px.line(one_row_df, x='data_date', y='price', color='to', title=plot_title)
p.update_layout(plot_bgcolor='white')
p.show()
```
